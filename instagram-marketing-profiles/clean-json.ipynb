```python
import json
import pandas as pd
import glob
import os
```

```python
pd.set_option("display.max_columns", 50)
```

```python
original_json_files = glob.glob("original/*.json")
os.makedirs("cleaned", exist_ok=True)
```

```python
data = []
for file in original_json_files:
    with open(file) as f:
        d = json.load(f)

        d.pop("id", None)
        d.pop("requestMetadata", None)
        d.pop("profilePicUrl", None)
        d.pop("profilePicUrlHD", None)
        d.pop("verified", None)

        for v in d.get("latestIgtvVideos", []):
            v.pop("id", None)
            v.pop("displayUrl", None)
            v.pop("ownerId", None)

            v_images = v.pop("images", [])
            v["imagesCount"] = len(v_images)

            for tu in v.get("taggedUsers", []):
                tu.pop("id", None)
                tu.pop("profilePicUrl", None)
                tu.pop("isVerified", None)

        for p in d.get("latestPosts", []):
            p.pop("id", None)
            p.pop("displayUrl", None)
            p.pop("ownerId", None)

            p_images = p.pop("images", [])
            p["imagesCount"] = len(p_images)

            p.pop("childPosts", [])

            for tu in p.get("taggedUsers", []):
                tu.pop("id", None)
                tu.pop("profilePicUrl", None)
                tu.pop("isVerified", None)

        for rp in d.get("relatedProfiles", []):
            rp.pop("id", None)
            rp.pop("profilePicUrl", None)
            rp.pop("isVerified", None)

        with open(f"cleaned-json-files/{os.path.basename(file)}", "w") as f_out:
            json.dump(d, f_out, indent=4)

        data.append(d)

# Create DataFrame
df = pd.json_normalize(data)

display(df.head(2))
print(df.shape)
```

# Mulitple Campustown Zillow Listings Cleaning

The dataset has been collected through [HasData](https://hasdata.com/), a web scraping service. The dataset contains over 30,000 listings of properties in the Campustowns across the United States.


````python
import pandas as pd
import numpy as np
````

The `filtered-listings.xlsx` contains result sets using this SQL query:

```sql
SELECT
  property_id,
  date_posted,
  days_on_zillow,
  price,
  last_sold_price,
  description,
  beds,
  baths,
  year_built,
  latitude,
  longitude,
  lot_size,
  living_area,
  address_street,
  address_city,
  address_state,
  address_zipcode,
  address_county,
  address_country,address_raw,
  county_fips,
  reso_is_new_construction,
  reso_architectural_style,
  reso_hoa_fee,
  reso_parking_capacity,
  reso_has_attached_garage
  reso_elementary_school,
  reso_middle_or_junior_school,
  reso_high_school
FROM zillow_properties;
```


````python
df = pd.read_excel("filtered-listings.xlsx")
df
````

Only filter cities with at least 200 listings. Sometimes, cities that are distant from campustowns are included in the search results, and we want to remove those cities from the dataset. After filtering, we have 28,000 listings in the dataset.


````python
cities_with_at_least_200_listings = (
    df["address_city"]
    .value_counts()
    .to_frame("count")
    .reset_index()
    .query("count > 200")
)
cities_with_at_least_200_listings
````

````python
df = df[
    df["address_city"].isin(cities_with_at_least_200_listings["address_city"])
].copy()
df.shape
````

Write as a parquet file.


````python
df.to_parquet("campustown_properties.parquet", index=False)
````
